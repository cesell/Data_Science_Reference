# Bar Chart

````python
layout = {
    "title":"Sales for first quarter",
    "xaxis": {"title":"Months", "tickangle":-20},
    "yaxis": {"title":"Units"}
}

trace = go.Bar(
    x = ['January','February','March'],
    y = [10, 11, 14],
    marker = dict(color = ['rgba(255,0,0,1)', 'rgba(204,204,204,1)','rgba(204,204,204,1)' ]),
    text = ["Below target", "Above target", "Above Target"]
)

data = [trace]

fig = go.Figure(data=data, layout=layout)

py.iplot(fig)
````

# Grouped Bar Charts

````python
trace0 = go.Bar(
    x = ['January','February','March'],
    y = [10, 11, 14],
    name = "Last Year"
)

trace1 = go.Bar(
    x = ['January','February','March'],
    y = [12, 13, 17],
    name = "This Year"
)

data = [trace0, trace1]

layout = {
    "title":"First quarter sales for two previous years",
    "xaxis": {"title":"Months"},
    "yaxis": {"title":"Units"},
    "barmode":"group"
}

py.iplot({"data":data, "layout":layout})
````

# Stacked Bar Chart

````python
trace0 = go.Bar(
    x = ['January','February','March'],
    y = [10, 11, 14],
    name = "Last Year"
)

trace1 = go.Bar(
    x = ['January','February','March'],
    y = [12, 13, 17],
    name = "This Year"
)

data = [trace0, trace1]

layout = {
    "title":"First quarter sales for two previous years",
    "xaxis": {"title":"Months"},
    "yaxis": {"title":"Units"},
    "barmode":"stack"
}

py.iplot({"data":data, "layout":layout})
````

# Frequency Distribution (normalized histogram)

````python
trace = go.Histogram(x=age, histnorm="probability")
data = [trace]
layout = {
    "title":"Frequency distribution of age",
    "xaxis":{"title":"Age in 5-year increments"}
         }
py.iplot({"data":data, "layout":layout})
````

# Two Datasets

````python
trace0 = go.Histogram(
    x = female.Age,
    name = "Female Age",
    xbins = dict(start=20, end=80, size=5)
)

trace1 = go.Histogram(
    x = male.Age,
    name = "Male Age",
    xbins = dict(start=20, end=80, size=5),
    opacity = 0.8
)

data = [trace0, trace1]

layout = {
    "title":"Age by gender",
    "xaxis": {"title":"Age in 5-year increments"},
    "barmode":"stack"
}

py.iplot({"data":data, "layout":layout})
````

# Cumulative Histogram

````python
trace = go.Histogram(
    x=df.Salary,
    cumulative=dict(enabled=True)
)

data = [trace]

layout = {"title":"Cumulative Histogram"}

py.iplot({"data":data, "layout":layout})
````

# Creating Data Point Values

Plotly's Python API contains a **figure factory** module which includes many wrapper functions that create unique chart types that are not yet included in plotly.js, Plotly's open-source graphing library. The figure factory functions create a full figure, so some Plotly features, such as subplotting, should be implemented slightly differently with these charts.

````python
fig = ff.create_distplot(
    hist_data = [df.Salary.values.tolist()],
    group_labels = ["Salary Distribution"],
    bin_size = [200]
)

fig["layout"].update(title="Salary Distribution")
py.iplot(fig)
````

# Distribution Plot of Two Datasets

````python
hist_data = [female.Salary.values.tolist(), male.Salary.values.tolist()]
group_labels = ["Female Salary", "Male Salary"]

fig = ff.create_distplot(hist_data, group_labels, bin_size=[200,200], colors=["rgba(20,20,20,0.8)", "rgba(150,150,150,0.8)"]) 

py.iplot(fig)
````

# Normal Curve

````python
hist_data = [female.Salary.values.tolist(), male.Salary.values.tolist()]
group_labels = ["Female Salary", "Male Salary"]

fig = ff.create_distplot(hist_data, group_labels, bin_size=[200,200], 
                         curve_type="normal",
                         show_hist=False,
                         show_rug=False 
                        )

fig.layout.update({"title":"Fitted normal curve", "xaxis":{"title":"Salary"}, "width":800, "height":600})

py.iplot(fig)
````

# Scatter Plot

````python
trace = go.Scatter(
    x=female_age,
    y=female_salary,
    mode="markers",
    marker=dict(size=12, color="rgba(255,70,0,0.9)")
)

data=[trace]

layout = go.Layout(
    title="Correlation between female age and salary",
    xaxis= dict(
                title="Age",
                zeroline=False
            ),
    height=600,
    width=800
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
````

# Adding more variable in a 2D Space

````python
trace = go.Scatter(
    x=female_age,
    y=female_salary,
    mode="markers",
    marker=dict(
        size=female_debt,
        color=female_tax,
        colorscale="Earth",
        showscale=True
    )
)

data = [trace]

layout = go.Layout(
    title="Correlation between female age and salary",
    xaxis=dict(title="Age", zeroline=False),
    yaxis=dict(title="Salary", zeroline=False),
    height=800,
    width=1000
)

fig=go.Figure(data=data, layout=layout)

py.iplot(fig)
````
