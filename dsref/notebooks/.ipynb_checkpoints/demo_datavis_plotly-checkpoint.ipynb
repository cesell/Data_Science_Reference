# An example to get excited

````python
import numpy as np

x = np.random.randn(2000)
y = np.random.randn(2000)

py.plot(
    [
        go.Histogram2dContour(x=x, y=y, contours=dict(coloring='heatmap')),
        go.Scatter(x=x, y=y, mode='markers', marker=dict(color='white',size=3,opacity=0.3))
    ],
    show_link=False
)
````

# Bar Chart

````python
layout = {
    "title":"Sales for first quarter",
    "xaxis": {"title":"Months", "tickangle":-20},
    "yaxis": {"title":"Units"}
}

trace = go.Bar(
    x = ['January','February','March'],
    y = [10, 11, 14],
    marker = dict(color = ['rgba(255,0,0,1)', 'rgba(204,204,204,1)','rgba(204,204,204,1)' ]),
    text = ["Below target", "Above target", "Above Target"]
)

data = [trace]

fig = go.Figure(data=data, layout=layout)

py.iplot(fig)
````

# Grouped Bar Charts

````python
trace0 = go.Bar(
    x = ['January','February','March'],
    y = [10, 11, 14],
    name = "Last Year"
)

trace1 = go.Bar(
    x = ['January','February','March'],
    y = [12, 13, 17],
    name = "This Year"
)

data = [trace0, trace1]

layout = {
    "title":"First quarter sales for two previous years",
    "xaxis": {"title":"Months"},
    "yaxis": {"title":"Units"},
    "barmode":"group"
}

py.iplot({"data":data, "layout":layout})
````

# Stacked Bar Chart

````python
trace0 = go.Bar(
    x = ['January','February','March'],
    y = [10, 11, 14],
    name = "Last Year"
)

trace1 = go.Bar(
    x = ['January','February','March'],
    y = [12, 13, 17],
    name = "This Year"
)

data = [trace0, trace1]

layout = {
    "title":"First quarter sales for two previous years",
    "xaxis": {"title":"Months"},
    "yaxis": {"title":"Units"},
    "barmode":"stack"
}

py.iplot({"data":data, "layout":layout})
````

# Histogram

````python
import random
import pandas as pd

#creating the data
np.random.seed(1234) #to get always the same values
age = np.random.uniform(low=21, high=75, size=100)
salary = np.random.normal(loc=3000, scale=1000, size=100)
gender = random.choices(["female","male"], k=100)

df = pd.DataFrame({"Age":age, "Salary":salary, "Gender":gender})
female = df[df.Gender == "female"]
male = df[df.Gender == "male"]
````

````python
trace = go.Histogram(x = age)
data = [trace]
py.iplot({"data":data})
````

# Frequency Distribution (normalized histogram)

````python
trace = go.Histogram(x=age, histnorm="probability")
data = [trace]
layout = {
    "title":"Frequency distribution of age",
    "xaxis":{"title":"Age in 5-year increments"}
         }
py.iplot({"data":data, "layout":layout})
````
